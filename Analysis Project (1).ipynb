## Questions

### 1.Find all record of the Surname each having CreditScore is more than 700 And Salary is more than 110000

```python
A=df['Surname'].loc[(df['CreditScore']>700) & (df['EstimatedSalary']>110000)].value_counts()
A
```

```python
# Barplot

A.plot(kind='bar',color='red')

plt.title("CreditScore records less than 700 and Estimeted salary 110000",color="black")

plt.show()
```

### 2.Find all records of Smith Surname person CustomerId and their CreditScore 

```python
df['CustomerId'].loc[(df['Surname']=='Smith') & (df['CreditScore'])].value_counts()
```

```python
df['CustomerId'].loc[(df['Surname']=='Smith') & (df['CreditScore'])].value_counts().plot(kind="bar")
plt.xlabel("Coustmer Id")
plt.title("Smith Coustmer CreditScore",color="Blue")
plt.show()
```

### 3.Find the records of Surname having EstimatedSalary is less than 120000 and CreditScore is less than 400

```python
c=df['Exited'].loc[(df['EstimatedSalary']<120000) & (df['CreditScore']<400)].mean()
c
```

### 4.Find all records of Thomas Surname custmers and their CustomerId And Loan Tenure

### 5.Find the records of Lin Surname,No of products & CreditScore details.

### 6.Find the record of Forbes Surname custmer Balance amout of loan and loan Tenure

### 7.Find the records of each custmers have loan Balance is more than 200000 and loan tenure less than 10 years.

### 8.Find the estimeted Salary and CreditScore of Mr.Loggia

```python
df[['EstimatedSalary','CreditScore']].loc[df['Surname']=='Loggia'].mean()
```

```python
df[['EstimatedSalary','CreditScore']].loc[df['Surname']=='Loggia'].mean().plot(kind='pie',autopct='%1.0f%%',explode=(0,0.3))

plt.title("Estimeted Salary And Creditscore Of Loggia Surname Coustmers")
plt.show()
```

### 9.Find the Records of McKay surname IsActive members and their creditscore and Exited from Loan

### 10.Find the records of surname they exited  from loan more than 3 and their tenure 10
